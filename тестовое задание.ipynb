```python
# 1. Во вкладке "Данные об аудитории" информация о пользователях, посетивших наше приложение в ноябре. Чему равен MAU продукта?
```

```python
import pandas as pd
```

```python
file = 'Данные для тестового задания - Данные об аудитории.csv'
df = pd.read_csv(file)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
november_data = df[(df['date'].dt.month == 11) & (df['date'].dt.year == 2023)]
```

```python
MAU = november_data['user_id'].nunique()
print(f'MAU ={MAU}')
```

```python
# 2. Используя вкладку "Данные об аудитории", посчитайте, чему будет равен DAU
```

```python
DAU_per_day = df.groupby('date')['user_id'].nunique()
```

```python
avg_DAU = DAU_per_day.mean()
print(f'DAU = {DAU_per_day.head()}')
print(f'average DAU = {round(avg_DAU)}')
```

```python
# 3. Используя вкладку "Данные об аудитории", посчитайте, чему будет равен retention первого дня у пользователей, пришедших в продукт 1 ноября
```

```python
cohort_date = pd.Timestamp('2023-11-01')
cohort_users = df[df['date'] == cohort_date]['user_id'].unique()
initial_count = len(cohort_users)
```

```python
return_date = pd.Timestamp('2023-11-02')
returning_users = df[(df['date'] == return_date) & (df['user_id'].isin(cohort_users))]['user_id'].nunique()
```

```python
retention_date = (returning_users / initial_count) * 100
print(f'1-ноября: {initial_count}')
print(f'вернулись на 1-день: {returning_users}')
print(f'retention 1-го дня: {retention_date:.2f}%')
```

```python
# 5. Во вкладке "Данные об аудитории" есть информация о том, сколько объявлений посмотрел каждый пользователь (view_adverts). 
# Посчитайте пользовательскую конверсию в просмотр объявления за ноябрь? (в пользователях) 
```

```python
# 18. Были получены следующие результаты. Коллеги просят вас подтвердить их и сделать окончательный вывод по эксперименту.
```

```python
import numpy as np
from scipy.stats import norm
```

```python
n_a, pay_a = 100047501, 1003
n_b, pay_b = 100001055, 1099
```

```python
cr_a = pay_a / n_a
cr_b = pay_b / n_b
```

```python
combined_cr = (pay_a + pay_b) / (n_a + n_b)
z_score = (cr_b - cr_a) / np.sqrt(combined_cr * (1- combined_cr) * (1/n_a + 1/n_b))
p_value = 2 * (1 - norm.cdf(abs(z_score)))
```

```python
lift = (cr_b- cr_a) / cr_a * 100
print(f'конверсия А: {cr_a:.7%}')
print(f'конверсия B: {cr_b:.7%}')
print(f'прирост: {lift:.2f}%')
print(f'Z статистика: {z_score:.4f}')
print(f'p-value: {p_value:.4f}')
```
